```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
# bar chart showing the count of tweets with different depression/stress scores: 

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/final_cleaned_tweets.csv')

score_counts = data.groupby('depression_stress_score')['Text'].count()
score_counts.plot(kind='bar')

plt.title('Count of Tweets by Depression/Stress Score')
plt.xlabel('Depression/Stress Score')
plt.ylabel('Count of Tweets')
plt.show()
```

```python
# histogram showing the distribution of vader sentiment scores:

vader_sentiment = data['vader_sentiment']

plt.hist(vader_sentiment)

plt.title('Distribution of Vader Sentiment Scores')
plt.xlabel('Vader Sentiment Score')
plt.ylabel('Count of Tweets')
plt.show()
```

```python
# scatter plot showing the relationship between depression/stress scores and vader sentiment scores

depression_stress_score = data['depression_stress_score']

plt.scatter(depression_stress_score, vader_sentiment)
plt.title('Relationship between Depression/Stress Scores and Vader Sentiment Scores')
plt.xlabel('Depression/Stress Score')
plt.ylabel('Vader Sentiment Score')
```

```python
# bar chart represening number of tweets in each stress level category
tweets_by_stress_level = data.groupby('stress_level_descriptions')['Text'].count()
tweets_by_stress_level.plot(kind='bar')

plt.title('Number of Tweets by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Number of Tweets')
plt.show()
```

```python
# bar chart showing the count of tweets with emoji_stress_scores

emoji_counts = data['emojie_stress_score'].value_counts()
ax = emoji_counts.plot(kind='bar')

plt.title('Count of Tweets by Emoji Stress Scores')
ax.set_xlabel('Emoji Stress Score')
ax.set_ylabel('Number of Tweets')
ax.yaxis.set_major_formatter('{x:,.0f}')
```

```python
# bar chart comparing the distribution of depression/stress scores for tweets with positive and negative vader sentiment scores
groups = data.groupby(data['vader_sentiment'] > 0)

pos_scores = groups.get_group(True)['depression_stress_score'].value_counts()
neg_scores = groups.get_group(False)['depression_stress_score'].value_counts()

ax = pd.concat([pos_scores, neg_scores], axis=1, sort=False).plot(kind='bar', width=0.4)
plt.title('Distribution of Depression/Stress Scores for Tweets with Positive and Negative Vader Sentiment Scores')
ax.set_xlabel('Depression/Stress Score')
ax.set_ylabel('Number of Tweets')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.legend(['Positive Vader Sentiment', 'Negative Vader Sentiment'])
plt.show()
```
