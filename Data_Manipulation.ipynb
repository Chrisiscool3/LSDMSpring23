## CLEAN TWEETS INTO PLAINTEXT WITHOUT SYMBOLS OR EMOJIS

```python
# define a function to remove emojis and special symbols from text
def clean_text(text):
    # remove emojis
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)
    # remove special symbols using regular expressions
    text = re.sub(r'[^\w\s]','', text)
    return text

# use the clean_text function to remove emojis and special symbols from each tweet
clean_tweets = [clean_text(tweet) for tweet in tweets]

# print the original tweets and the cleaned tweets side by side
for i in range(len(tweets)):
    print(clean_tweets[i])
```

## CREATE LIST OF ENGLISH TWEETS ONLY

```python
def is_english(text):
    try:
        lang = detect(text)
        if lang == 'en':
            return True
        else:
            return False
    except:
        return False

english_tweets = [tweet for tweet in clean_tweets if is_english(tweet)]
```

## STRESS LEVEL TEXT-INDICATOR

```python
import pandas as pd

#Read CSV into dataframe
#EDIT TO YOUR CORRECT CSV NAME
df = pd.read_csv('chris_cleaned.csv')

#drop any rows that contain NaN
df.dropna(inplace=True)

df.head()
```

```python
#Remove any rows that contain non-numerical data, every row should have numerical data in the scoring column
df = df[pd.to_numeric(df['depression_stress_score'], errors='coerce').notnull()]
```

```python
#list of score values as strings
stress_levels = df['depression_stress_score']
# Convert to list of integers
int_levels = [int(x) for x in stress_levels]

print(min(int_levels))
print(max(int_levels))
```

```python
#dictionary to identify each level of stress
stress_descriptions = {
    -5: 'Extremely low',
    -4: 'Very low',
    -3: 'Low',
    -2: 'Somewhat low',
    -1: 'Slightly low',
    0: 'Neutral',
    1: 'Slightly high',
    2: 'Somewhat high',
    3: 'High',
    4: 'Very high',
    5: 'Extremely high'
}

#create a list and append to that list all the decription values of each level
stress_desc = []
for level in int_levels:
  stress_desc.append(stress_descriptions[level])
```

```python
#create a new column in the dataframe that has the level descriptions
df['stress_level_descriptions'] = stress_desc
```
